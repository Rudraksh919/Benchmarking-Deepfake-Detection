```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from classifiers import Meso4, XceptionClassifier
from tensorflow.keras.applications.xception import preprocess_input
import tensorflow as tf
```

```python
meso=model = Meso4()
Xception = XceptionClassifier()
```

```python
datagen_meso = ImageDataGenerator(validation_split=0.2)

train_gen_meso = datagen_meso.flow_from_directory(
    'train_faces',
    target_size=(256, 256),
    batch_size=32,
    class_mode='binary',
    subset='training'
)

val_gen_meso = datagen_meso.flow_from_directory(
    'train_faces',
    target_size=(256, 256),
    batch_size=32,
    class_mode='binary',
    subset='validation'
)

datagen_xception = ImageDataGenerator(preprocessing_function=preprocess_input, validation_split=0.2)

train_gen_xception = datagen_xception.flow_from_directory(
    'train_faces',
    target_size=(256, 256),
    batch_size=8,
    class_mode='binary',
    subset='training'
)

val_gen_xception = datagen_xception.flow_from_directory(
    'train_faces',
    target_size=(256, 256),
    batch_size=8,
    class_mode='binary',
    subset='validation'
)
```

```python
meso.model.fit(train_gen_meso, validation_data=val_gen_meso, epochs=30)
```

```python
meso.model.save_weights("Custom-Trained-Weights/MesoNet.weights.h5")
```

```python
tf.keras.backend.clear_session()
```

```python
Xception.model.fit(train_gen_xception, validation_data=val_gen_xception, epochs=30)
```

```python
Xception.model.save_weights("Custom-Trained-Weights/XceptionNet.weights.h5")
```

```python
tf.keras.backend.clear_session()
```
